```python
import torch
import numpy as np
from sklearn.manifold import TSNE
```

```python
cluster_all = torch.load('./cluster_all_data.pt')
print(cluster_all.shape)
```

```python
a_cluster_all = torch.flatten(cluster_all, start_dim=1)
a_cluster_all.shape
```

```python
tsne = TSNE(n_components=2, random_state=1)
X_2d = tsne.fit_transform(a_cluster_all.cpu().numpy())
```

```python
X_2d = X_2d/np.max(np.abs(X_2d))
```

```python
from matplotlib import pyplot as plt

plt.figure(figsize=(6, 5), dpi=200)
plt.scatter(X_2d[:,0],X_2d[:,1], s=1)
plt.show()
```
